1. Write a function to calculate the number of words, number of lines, and length of a string the same way the wccommand does in the command line

```python
def wc(x):
    chars=0
    for char in x:
        chars += 1
    words = len(x.split(" "))
    lines = len(x.split("\n"))
    return (lines, words, chars)
```

__list.append(x)__

Add an item to the end of the list. Equivalent to a[len(a):] = [x].

__list.extend(iterable)__

Extend the list by appending all the items from the iterable. Equivalent to a[len(a):] = iterable.

List concatenation (with `+`) is a more expensive operation than extend(), since a new list must be created and the objects copied over. Using extend to append elements to an existing list is preferable especially if you are building up a large list.

__list.insert(i, x)__

Insert an item at a given position.

__list.remove(x)__

Remove the first item from the list whose value is equal to x. It raises a ValueError if there is no such item.

__list.pop([i])__

Remove the item at the given position in the list, and return it. If no index is specified, a.pop() removes and returns the last item in the list.

__list.clear()__

Remove all items from the list. Equivalent to del a[:].

__list.index(x[, start[, end]])__

Return zero-based index in the list of the first item whose value is equal to x. Raises a ValueError if there is no such item.

 

__list.count(x)__

Return the number of times x appears in the list.

__list.sort(key=None, reverse=False)__

Sort the items of the list in place (the arguments can be used for sort customization, see sorted() for their explanation).

__list.reverse()__

Reverse the elements of the list in place.

__list.copy()__

Return a shallow copy of the list. Equivalent to a[:].

## List comprehensions

List comprehensions provide a concise way to create lists. Common applications are to make new lists where each element is the result of some operations applied to each member of another sequence or iterable.

Allows to concisely form a new list by filtering the elements of a collection and transforming the elements passing the filter in one concise expression.

```python
squares = []
squares = list(map(lambda x: x**2, range(10)))
squares
```

Or, equivalently


```python
squares = [x**2 for x in range(10)]
```

```python
[(x, y) for x in [1,2,3] for y in [3,1,4] if x != y]
[(1, 3), (1, 4), (2, 3), (2, 1), (2, 4), (3, 1), (3, 4)]
```

```python
#which is equal to
combs = []
for x in [1,2,3]:
    for y in [3,1,4]:
        if x != y:
            combs.append((x, y))
combs
[(1, 3), (1, 4), (2, 3), (2, 1), (2, 4), (3, 1), (3, 4)]
```

```python
a=[1,2,3,4, 5, 6]
result = []
for val in a:
    if val%2:
        result.append(val**2)
result
```

## Nested list comprehensions

The initial expression in a list comprehension can be any arbitrary expression, including another list comprehension.

* nested list comprehensions are a bit hard to wrap your head around. 
* The for parts of the list comprehension are arranged according to the order of nesting, 
* filter condition is put at the end as before. 
* example
  * we flatten a list of tuples of integers into a simple list of integers:


```python
some_tuples=[(1,2,3),(4,5,6),(7,8,9)]
flatten=[]
for tup in some_tuples:
    for element in tup:
        flatten.append(element)
flatten
```

```python
some_tuples=[(1,2,3),(4,5,6),(7,8,9)]
#[ return_element for_loops_for_coming_to_element ]
flatten2=[element for tup in some_tuples for element in tup]
flatten2
```

4. Reverse word order from the input string

```python
def reverse(x):
    y = x.split()
    result = []
    for word in y:
        result.insert(0,word)
    return " ".join(result)
```

```python
reverse("est ya y frase una es esto")
```

6. Write a Python function that takes a list of words and returns the length of the longest 

```python
def find_longest_word(words_list):
    word_len = []
    for i in words_list:
        word_len.append((len(i),i))
    word_len.sort()
    return word_len[-1][1]
find_longest_word(('pablo', 'peddvsvsdvsdvsdro', 'osc', 'eldcdvefante', 'pp'))
```

```python
def find_longest_word(words_list):
    return len(max(words_list, key = len))

names = ["Bob", "Igor", "Anita", "Israel", "Thomas", "Bill", "Ronaldo", "Messi"] 
print(find_longest_word(names))
```
