## Output Formatting

If you want to use formatted string literals, begin with `f` or `F` before the opening quotation mark.

```python
sport = "mixed martial arts"
year = 2020
f'He was the champion of {sport} in {year}'
```

str() returns representations of values fairly human-readable

```python
s = "Hello my Name is Eminem"
str(s)
```

## Formatted string literals

f-strings let you include the value of Python expressions inside a string by prefixing the string with `f``or `F`

```python
import math
print(f'The value of pi is approximately {math.pi:.3f}.')
```

Passing an integer after the ':' will cause that field to be a minimum number of characters wide. This is useful for making columns line up.

```python
table = {'Ronaldo': 26, 'Messi': 29, 'Neymar': 22}
for name, phone in table.items():
    print(f'{name:10} ==> {phone:10d}')
```

## The String format () method

The brackets and characters are replaced with the objects passed into the str.format()

```python
print('{0} and {1}'.format('spam', 'eggs'))
```

*The Python programming language, named after Monty Python, prefers to use spam, ham, and eggs as metasyntactic variables, instead of the traditional foo, bar and baz.*

## Manual String Formatting

```python
for x in range(1,11):
    print(repr(x).rjust(2), repr(x*x).rjust(3), end = '  XX ')
    print(repr(x*x*x).rjust(4))
```

2. Open Finn.txt  file, read lines into a list. Create a new version of Finn_nbl.txt with no blank lines.

```python
path = '/home/dsc/Data/shell/Finn.txt'
f = open(path)
lines = []
for line in f:
    lines.append(line)
f.close()
len(lines)

f_out=open("Finn_nbl.txt", 'w')
for line in lines:
    if len(lines)>2:
        f_out.writelines(line)
f_out.close()
```
