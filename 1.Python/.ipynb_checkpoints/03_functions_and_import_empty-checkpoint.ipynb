# Funciones

Las definimos con `def` y devuelven un único valor.

##### Si una función no tiene un `return`, siempre devuelve `None`

```python
def distance(point1, point2):
    x=point2[0] - point1[0]
    y=point2[1] - point1[1]
    return (x**2 +y**2) **0.5
```

```python
distance([45,25],[11,3])
```

## Docstrings

```python
def distance(point1, point2):
    ''' Returns Euclid
    distance
    point1 and point2 should be 2D points coordiantes
    ex. point1=[3,1];point2=[2,3]
    '''
    
    x=point2[0] - point1[0]
    y=point2[1] - point1[1]

    return (x**2 +y**2) **0.5
```

```python
help(distance)
```

```python
def ojoCuidao(point1, point2):
    ''' Returns Euclid
    distance
    point1 and point2 should be 2D points coordinates
    ex. point1=[3,1];point2=[2,3]
    '''
    
    x=point2[0] + point1[0]
    y=point2[1] + point1[1]

    return x,y
```

```python
ojoCuidao("pp","kk")
```

```python
def max_of_three(a,b,c):
    max=a
    if b > a and b > c:
        max = b
    elif c > b and c > a:
        max = c
    return max
```

```python
max_of_three(111,54,333)
```

```python
def max_of_three(a,b,c):
    max_total=0
    #STEP 1 => CHECK IF ELEMENT 0 is BIGGER THAN MAX
            #if num>max_total:
            #   max_total=num
    if a>max_total:
        max_total = a
    #STEP 2 => CHECK IF  ELEMENT 1 is BIGGER THAN MAX
    if b>max_total:
        max_total = b
    #STEP 3 => CHECK IF ELEMENT 2 is BIGGER THAN MAX
    if c>max_total:
        max_total=c
    return max_total
```

```python
def max_list(list1):
    max_total=list1[0]
    for num in list1:
        if num>max_total:
            max_total=num
    return max_total
    
```

```python
list1 = [0,11,3,21,3,2,4,6,7,5,3,5,67,3,22,4,54]
max_list(list1)
```

```python
def centenario(name,year):
    year_int=int(year)
    if isinstance(year_int, int):
        print(f"{name} will reach 100 years in {year_int + 100}")
```
